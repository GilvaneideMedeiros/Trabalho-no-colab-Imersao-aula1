# Aula 1 - Imersão Dados com Python

```
import pandas as pd
```

```
df = pd.read_csv("https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv")
```

```
df.head(10)
```

```
df.info()
```

```
df.describe()
```

```
df.shape
```

```
linhas, colunas = df.shape[0], df.shape[1]
print('Nº de linhas:', linhas)
print('Nº de colunas:', colunas)
```

```
df.columns
```

```
renomear_colunas = {
    'ano': 'ano',
    'nivel_experiencia': 'senioridade',
    'tipo_emprego': 'contrato',
    'cargo': 'cargo',
    'salario': 'salario',
    'moeda_salario': 'moeda',
    'salario_em_usd': 'usd',
    'residencia_funcionario': 'residencia',
    'taxa_remoto': 'remoto',
    'localizacao_empresa': 'empresa',
    'porte_empresa': 'tamanho_empresa'
}

df.rename(columns=renomear_colunas, inplace=True)
df.columns
```

```
df['remoto'].value_counts()
```

```
df['contrato'].value_counts()
```

```
df['senioridade'].value_counts()
```

```
df['tamanho_empresa'].value_counts()
```

```
senioridade = {
    'SE': 'Sênior',
    'MI': 'Médio',
    'EN': 'Entrada',
    'EX': 'Executivo'
}

df['senioridade'] = df['senioridade'].replace(senioridade)
df['senioridade'].value_counts()
```

```
print('Primeiras 10 linhas do DataFrame com senioridade traduzida:')
display(df.head(10))
```

```
contrato = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
}

df['contrato'] = df['contrato'].replace(contrato)
display(df['contrato'].value_counts())
```

```
print('Primeiras 10 linhas do DataFrame com a coluna \'contrato\' traduzida:')
display(df.head(10))
```

```
tamanho_empresa = {
    'M': 'Média',
    'L': 'Grande',
    'S': 'Pequena'
}

df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa)
display(df['tamanho_empresa'].value_counts())
```

```
print('Primeiras 10 linhas do DataFrame com a coluna \'tamanho_empresa\' traduzida:')
display(df.head(10))
```

```
print('Categorias originais da coluna \'remoto\':')
display(df['remoto'].value_counts())
```

```
remoto = {
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
}

df['remoto'] = df['remoto'].replace(remoto)
df['remoto'].value_counts()
```

```
df.head()
```

```
print('Primeiras 10 linhas do DataFrame com a coluna \'remoto\' traduzida:')
display(df.head(10))
```

```
print('Cargos mais frequentes na coluna "cargo":')
display(df['cargo'].value_counts().head(20))
```

```
cargo = {
    'Data Engineer': 'Engenheiro de Dados',
    'Data Scientist': 'Cientista de Dados',
    'Machine Learning Engineer': 'Engenheiro de Machine Learning',
    'Analytics Engineer': 'Engenheiro de Analytics',
    'Research Scientist': 'Cientista Pesquisador',
    'Data Analyst': 'Analista de Dados',
    'ML Engineer': 'Engenheiro de ML',
    'Applied Scientist': 'Cientista Aplicado',
    'Data Architect': 'Arquiteto de Dados',
    'Software Engineer': 'Engenheiro de Software',
    'Research Engineer': 'Engenheiro de Pesquisa',
    'Data Science Manager': 'Gerente de Ciência de Dados',
    'Cloud Engineer': 'Engenheiro de Nuvem',
    'MLOps Engineer': 'Engenheiro de MLOps',
    'Prompt Engineer': 'Engenheiro de Prompt',
    'AI Engineer': 'Engenheiro de IA',
    'Business Intelligence Engineer': 'Engenheiro de Business Intelligence',
    'Head of Data Science': 'Diretor de Ciência de Dados',
    'Data Product Manager': 'Gerente de Produto de Dados',
    'Machine Learning Scientist': 'Cientista de Machine Learning'
}

df['cargo'] = df['cargo'].replace(cargo)
df['cargo'].value_counts().head(20)
```
